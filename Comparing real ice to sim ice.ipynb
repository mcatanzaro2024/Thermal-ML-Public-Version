```python
#stainless
HeatSimulation.SetSize(.005,.025)
HeatSimulation.SetNodes(8,.155)
HeatSimulation.SetMaterialProperties(7800,500,16.2)
HeatSimulation.SetConvectionProperties(4,23)
HeatSimulation.EnableSurfaceHeatTransfer(False)
HeatSimulation.SetHeatSource(0)
HeatSimulation.SetInitalTemp(25)
HeatSimulation.SetTimeIncrement(10)
```

```python
Desmos = pd.read_csv('tblIceOnSteel.csv')
#Desmos
```

```python
HeatSimulation.EnableSurfaceHeatTransfer(True)
HeatSimulation.SetHeatSource(0)
HeatSimulation.SetInitalTemp(23)
HeatSimulation.GetTempatureAtNodeAtTime(0,6000,4)
```

```python
plt.plot(Desmos.iloc[:,0],Desmos.iloc[:,4], label="DESMOS")
Trial = HeatSimulation.AllTemp
Node = 0
while Node < HeatSimulation.NumNodes:
    Node = Node + 2
    TrialValue = Node + 1
    LabelValue = "Node" + str(Node)
    plt.plot(Trial.Time, Trial.iloc[:,TrialValue], label=LabelValue)
plt.plot(Trial.Time, Trial.Heater, label="HeatSource")
plt.plot(Trial.Time, Trial.Convection, label="Convection")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.legend()
plt.show()
```

```python
HeatSimulation.AllTemp
```

```python
T = 0
while T < 23:
    T = T + 1
    Line = T
    TimeSim = Desmos.loc[Line,'time']
    Temp = Desmos.iloc[Line,1:5].to_numpy()
    X = (0,50,100,155)
    #HeatSimulation.SetTimeIncrement(TimeSim/(Desmos.loc[len(Desmos)-1,'time']))
    HeatSimulation.GetTempatureAtNodeAtTime(0,TimeSim,4)
    MyTemp = HeatSimulation.AllTemp.iloc[(len(HeatSimulation.AllTemp)-1),1:10].to_numpy()
    MyX = [0,1*19.375,2*19.375,3*19.375,4*19.375,5*19.375,6*19.375,7*19.375,8*19.375]
    plt.plot(X,Temp, label="Desmos",marker= "h")
    plt.plot(MyX,MyTemp, label="MyModel",marker= "p")
    plt.xlabel("Distance(mm)")
    plt.ylabel("Temperature (C)")
    plt.title('Time ' + str(TimeSim))
    plt.ylim(0, 25)
    plt.legend()
    plt.show()
```
